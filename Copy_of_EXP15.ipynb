```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,4500],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount': [10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Price Normalization
min_price = df['Price'].min()
max_price = df['Price'].max()

df['Price_Normalized'] = (df['Price'] - min_price) / (max_price - min_price)

print(df[['Product', 'Price', 'Price_Normalized']])
```

```python
#Normalization of Multiple Columns
cols = ['Price', 'Units_Sold', 'Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
df_norm
```

```python
#Z-Score Normalization
df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
df
```

```python
#Decimal Scaling
df['Price_Scaled'] = df['Price'] / 10000

print(df[['Product','Price', 'Price_Scaled' ]])
```

```python
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df1 = pd.DataFrame(data)
df1
```

```python
#Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender_Label'] = le.fit_transform(df1['Customer_Gender'])
df1
```

```python
df1['City_Label'] = le.fit_transform(df1['City'])
df1
```

```python
#Method 2: One-Hot Encoding
df_encoded = pd.get_dummies(df1, columns=['Customer_Gender'])
df_encoded
```

```python
df_dummies = pd.get_dummies(df1, columns=['Payment_Method'], drop_first=True)
df_dummies
```
